```python
from pyexpat import features
from sklearn.tree import _tree

feature_names = ['soil1', 'soil2', 'temp', 'humidity', 'light']

def emit_rules_as_cpp(tree, feature_names):
    tree_ = tree.tree_
    features = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    def recurse(node, indent=0):
        pad = "  " * indent
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = features[node]
            thresh = tree_.threshold[node]
            print(f"{pad}if (x[{feature_names.index(name)}] <= {thresh:.6f}) {{")
            recurse(tree_.children_left[node], indent + 1)
            print(f"{pad}}}")
        else:
            cls = tree_.value[node].argmax()
            print(f"{pad}return {cls};")

    recurse(0, 0)
```

```python
import joblib

baseline_decision_tree = joblib.load("models/baseline_decision_tree.joblib")

emit_rules_as_cpp(baseline_decision_tree, feature_names)
```
